```python
import pandas as pd

train_df = pd.read_csv("train.csv")
test_df = pd.read_csv("test.csv")
```

```python
categorical_cols = train_df.select_dtypes(exclude=["number"]).columns
```

```python
train_df_dropped = train_df.drop(categorical_cols, axis=1)
train_df_dropped
```

```python
train_df[categorical_cols].fillna(" ", inplace=True)
test_df[categorical_cols].fillna(" ", inplace=True)
```

```python
from sklearn.preprocessing import OrdinalEncoder

ordinal_encoder = OrdinalEncoder()
train_df_ordinal_encoded = train_df.copy()

train_df_ordinal_encoded[categorical_cols] = pd.DataFrame(ordinal_encoder.fit_transform(train_df[categorical_cols]), columns = categorical_cols)
train_df_ordinal_encoded
```
