# Creating the dataset and model
This is what the dataset will look like. We are trying to use horsepower and seats to predict the label.
Screenshot 2021-12-01 8.39.58 PM.png


```python
#Import
from sklearn import tree
# Create features and labels
features = [[300, 2], [450, 2], [200, 8], [150, 9]]
labels = ["sports-car", "sports-car", "minivan", "minivan"]

# Create model

model = tree.DecisionTreeClassifier()
model.fit(features, labels)
```

### Testing the model
image.png

```python
print(model.predict([[500, 2], [100, 10], [150, 6]]))
```

# Using a real world dataset
https://www.kaggle.com/c/titanic/data?select=train.csv

Features to use:
* Pclass - Ticket class
* SibSp -# of siblings / spouses aboard the Titanic
* Fare - Passenger Fare

Label- Survived

```python
import pandas as pd
from sklearn import tree

train_df = pd.read_csv('train.csv')
```

```python
print(train_df.head())
feature_columns = ['Pclass', 'SibSp']
```

```python
train_X = train_df[feature_columns]
train_y = train_df['Survived']
```

```python
clf = tree.DecisionTreeClassifier()
clf.fit(train_X, train_y)
```

```python
test_df = pd.read_csv('test.csv')
test_X = test_df[feature_columns]
```

```python
print(clf.predict(test_X))
```
